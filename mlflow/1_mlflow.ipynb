# MLFLOW

Dataset → Train Model → Log Params → Log Metrics → Save Model



# Step 1 — MLflow Run ka Structure

Ek MLflow experiment usually aise hota hai:

```text
Start Run
   ↓
Log Parameters
   ↓
Train Model
   ↓
Log Metrics
   ↓
Save Model (Artifact)
End Run
```

```text
Training Script
      ↓
MLflow Tracking
      ↓
MLflow Server
      ↓
MLflow UI
End Run
```

````python
import mlflow
````

````python
with mlflow.start_run():
    print("MLflow run started")
````

````python
mlflow.log_param('max_depth',10)
mlflow.log_param('n_estimators', 100)
````

````python
mlflow.log_metric('accuracy',0.87)
````
